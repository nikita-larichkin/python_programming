```python
phone_book = {}


def standardize_phone(phone: str) -> str:
    phone_digits = ''.join(filter(str.isdigit, phone))
    standardized_phone = "+7" + phone_digits[-10:]
    return standardized_phone


def standardize_name(name: str) -> str:
    words = name.split()
    standardized_name = ' '.join([word.capitalize() for word in words])
    return standardized_name


def add_contact() -> None:
    name = input("Введите имя контакта: ")
    phone = input("Введите номер телефона: ")

    standardized_name = standardize_name(name)
    standardized_phone = standardize_phone(phone)

    phone_book[standardized_name] = standardized_phone

    print("Контакт успешно добавлен!")


def delete_contact() -> None:
    name = input("Введите имя контакта для удаления: ")

    standardized_name = standardize_name(name)

    if standardized_name in phone_book:
        del phone_book[standardized_name]
        print("Контакт успешно удален!")
    else:
        print("Контакт не найден!")


def view_phone_book() -> None:
    if len(phone_book) > 0:
        print("--- Телефонная книга ---")
        for item in phone_book.items():
            print(*item)
    else:
        print("Телефонная книга пуста!")


def change_phone_number() -> None:
    name = input("Введите имя контакта, у которого нужно изменить номер телефона: ")

    standardized_name = standardize_name(name)

    if standardized_name in phone_book:
        phone = input("Введите новый номер телефона: ")

        standardized_phone = standardize_phone(phone)

        phone_book[standardized_name] = standardized_phone

        print("Номер телефона успешно изменен!")
    else:
        print("Контакт не найден!")


while True:
    print("\n--- Телефонная книга ---")
    print("1. Добавить контакт")
    print("2. Удалить контакт")
    print("3. Просмотреть телефонную книгу")
    print("4. Изменить номер телефона")
    print("0. Выход")

    choice = input("Выберите действие: ")

    match choice:
        case "1":
          add_contact()
        case "2":
          delete_contact()
        case "3":
          view_phone_book()
        case "4":
          change_phone_number()
        case "0":
          print("Выход из программы...")
          break

        case _:
          print("Некорректный выбор.")
```
